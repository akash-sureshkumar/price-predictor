### Prediction

Now that the models are trained, I'll generate future dates for the next 6 months and predict the gold and silver prices.

```python
# Create a dataframe with future dates for the next 6 months
future_period_months = 6
future = gold_model.make_future_dataframe(periods=future_period_months * 30, include_history=False) # Approximate days in 6 months

# We need to add the regressors for the future dates.
# For demonstration purposes, I'll use a simple approach by extending the last known values.
# In a real-world scenario, you would need to forecast these regressors as well.
last_inflation = df['Inflation_Rate_India'].iloc[-1]
last_exchange_rate = df['USD_INR_Exchange_Rate'].iloc[-1]

future['Inflation_Rate_India'] = last_inflation
future['USD_INR_Exchange_Rate'] = last_exchange_rate


# Predict future Gold prices
gold_forecast = gold_model.predict(future)

# Predict future Silver prices
silver_forecast = silver_model.predict(future)

# Combine the predictions into a single dataframe
predicted_prices_df = future.copy()
predicted_prices_df['Predicted_Gold_Price_INR'] = gold_forecast['yhat']
predicted_prices_df['Predicted_Silver_Price_INR'] = silver_forecast['yhat']

# Rename the 'ds' column to 'Date' for clarity
predicted_prices_df.rename(columns={'ds': 'Date'}, inplace=True)

# Display the predicted prices
print("Predicted Gold and Silver Prices for the next 6 months:")
display(predicted_prices_df.head())
```

### Visualization of Predictions

I'll visualize the predicted Gold and Silver prices along with the historical data to see the trend.

```python
import matplotlib.pyplot as plt

# Plot historical and predicted Gold prices
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Gold_Price_INR'], label='Historical Gold Price (INR)', color='gold')
plt.plot(predicted_prices_df['Date'], predicted_prices_df['Predicted_Gold_Price_INR'], label='Predicted Gold Price (INR)', color='orange', linestyle='--')
plt.title('Gold Price Prediction for the next 6 Months')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()

# Plot historical and predicted Silver prices
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Silver_Price_INR'], label='Historical Silver Price (INR)', color='silver')
plt.plot(predicted_prices_df['Date'], predicted_prices_df['Predicted_Silver_Price_INR'], label='Predicted Silver Price (INR)', color='gray', linestyle='--')
plt.title('Silver Price Prediction for the next 6 Months')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()
```
